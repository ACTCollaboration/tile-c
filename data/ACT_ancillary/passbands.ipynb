```python
#For PA1/2/3, determine upper and lower limits where to cut off the integrals by finding where the responses
#  are consistent with zero [or negative] (within the uncertainties) and then going 5 GHz beyond that.
def find_cutoffs(nu,fnu,fnu_err,buff=5.): #buff = 5 GHz
    N_nu = len(nu)
    for i in range(N_nu-4):
        if (fnu[i]-fnu_err[i] <= 0. and fnu[i+1]-fnu_err[i+1] > 0. and fnu[i+2]-fnu_err[i+2] > 0. and fnu[i+3]-fnu_err[i+3] > 0. and fnu[i+4]-fnu_err[i+4] > 0.):
            lower_cutoff = nu[i]
            break
    for i in range(4,N_nu):
        if (fnu[i]-fnu_err[i] <= 0. and fnu[i-1]-fnu_err[i-1] > 0. and fnu[i-2]-fnu_err[i-2] > 0. and fnu[i-3]-fnu_err[i-3] > 0. and fnu[i-4]-fnu_err[i-4] > 0.):
            upper_cutoff = nu[i]
            break
    print lower_cutoff, upper_cutoff
    lower_cutoff -= buff
    upper_cutoff += buff
    if (lower_cutoff < nu[0]):
        lower_cutoff = nu[0]
    if (upper_cutoff > nu[-1]):
        upper_cutoff = nu[-1]
    print lower_cutoff, upper_cutoff
    # find elements in nu array nearest to the cutoffs
    idx_low = (np.abs(nu - lower_cutoff)).argmin()
    idx_high = (np.abs(nu - upper_cutoff)).argmin()
    # return the three truncated arrays
    return nu[idx_low:idx_high+1],fnu[idx_low:idx_high+1],fnu_err[idx_low:idx_high+1]

print "PA1:"
nu_PA1_cut, fnu_PA1_cut, fnu_err_PA1_cut = find_cutoffs(nu_PA1, fnu_PA1, fnu_err_PA1)
print nu_PA1_cut[0],nu_PA1_cut[-1]
print "PA2:"
nu_PA2_cut, fnu_PA2_cut, fnu_err_PA2_cut = find_cutoffs(nu_PA2, fnu_PA2, fnu_err_PA2)
print nu_PA2_cut[0],nu_PA2_cut[-1]
print "PA3_150:"
nu_PA3_150_cut, fnu_PA3_150_cut, fnu_err_PA3_150_cut = find_cutoffs(nu_PA3_150, fnu_PA3_150, fnu_err_PA3_150)
print nu_PA3_150_cut[0],nu_PA3_150_cut[-1]
print "PA3_90:"
nu_PA3_90_cut, fnu_PA3_90_cut, fnu_err_PA3_90_cut = find_cutoffs(nu_PA3_90, fnu_PA3_90, fnu_err_PA3_90)
print nu_PA3_90_cut[0],nu_PA3_90_cut[-1]

# save truncated versions
np.savetxt('PA1_avg_passband_wErr_trunc.txt', np.transpose(np.array([nu_PA1_cut, fnu_PA1_cut, fnu_err_PA1_cut])))
np.savetxt('PA2_avg_passband_wErr_trunc.txt', np.transpose(np.array([nu_PA2_cut, fnu_PA2_cut, fnu_err_PA2_cut])))
np.savetxt('PA3_avg_passband_90_wErr_trunc.txt', np.transpose(np.array([nu_PA3_90_cut, fnu_PA3_90_cut, fnu_err_PA3_90_cut])))
np.savetxt('PA3_avg_passband_150_wErr_trunc.txt', np.transpose(np.array([nu_PA3_150_cut, fnu_PA3_150_cut, fnu_err_PA3_150_cut])))
```

```python
### PA2: Effective central frequency to compact sources ###

#For ACT, al is near -1
al_1 = 0.0 #use average of al=-1 and al=0 for compact sources??
al_2 = -1.0 #use average of al=-1 and al=0 for compact sources??
be_s = -0.7 
be_ff = -0.1
be_d = 3.7 
be_RJ = 2.0 

nu = np.loadtxt('PA2_avg_passband_wErr.txt')[36:161,0] #cutting off below 118 GHz
fnu = np.loadtxt('PA2_avg_passband_wErr.txt')[36:161,1] #cutting off above 180 GHz
print np.min(nu), np.max(nu)
dnu = nu[1]-nu[0]
nu *= 1.e9 #JCH fix: was missing factor of 1e9 (GHz -> Hz)
dnu *= 1.e9 #JCH: convert to Hz

kB = 1.38e-23
c = 3e8 #m/s
h = 6.625e-34
TCMB = 2.726 #JCH (Rahul was using 2.732, not sure why)
x = h*nu/kB/TCMB
e = 2.718

nu_be_CMB = 2.*(kB*TCMB)**3. / (h*c)**2. * e**x * x**4 / (e**x - 1)**2 #JCH: include prefactor
nu_be_SZ = nu_be_CMB * (x * 1./np.tanh(x/2.) - 4.) #JCH: fix this (it was missing the nu_be_CMB factor)

#JCH
nu_test = (np.linspace(1.,1000.,num=1000))*1.e9
x_test = h*nu_test/kB/TCMB
nu_be_CMB_test = 2.*(kB*TCMB)**3. / (h*c)**2. * e**x_test * x_test**4 / (e**x_test - 1)**2
nu_be_SZ_test = nu_be_CMB_test * (x_test * 1./np.tanh(x_test/2.) - 4.)
plt.figure(figsize=(10,6))
plt.plot(nu_test, nu_be_CMB_test, color='k', label='CMB')
plt.plot(nu_test, nu_be_SZ_test, color='b', label='tSZ', ls='--')
plt.tick_params(labelsize=20) 
plt.xlabel(r'$\nu$ [Hz]', fontsize=20)
plt.ylabel(r'$\Delta I(\nu)$ [W/m${}^2$/Hz]', fontsize=20)
plt.legend(loc='upper left', fontsize=14)
plt.show()

print 'Average of al=0 and al=-1 for compact sources (for ACT, al is closer to -1):'
nu_eff_s1 = np.sum(nu * fnu * nu**(-2.*(1.+al_1)) * nu**(be_s) * dnu)/np.sum(fnu * nu**(-2.*(1.+al_1)) * nu**(be_s) * dnu)
nu_eff_s2 = np.sum(nu * fnu * nu**(-2.*(1.+al_2)) * nu**(be_s) * dnu)/np.sum(fnu * nu**(-2.*(1.+al_2)) * nu**(be_s) * dnu)
nu_eff_s = 0.5*(nu_eff_s1+nu_eff_s2)
print 'nu_eff_synch =', "{0:.1f}".format(nu_eff_s/1.e9), '+/-', 2.4, 'GHz' #based on statistical and systematic error estimates from the FTS measurement

nu_eff_ff1 = np.sum(nu * fnu * nu**(-2.*(1.+al_1)) * nu**(be_ff) * dnu)/np.sum(fnu * nu**(-2.*(1.+al_1)) * nu**(be_ff) * dnu)
nu_eff_ff2 = np.sum(nu * fnu * nu**(-2.*(1.+al_2)) * nu**(be_ff) * dnu)/np.sum(fnu * nu**(-2.*(1.+al_2)) * nu**(be_ff) * dnu)
nu_eff_ff = 0.5*(nu_eff_ff1+nu_eff_ff2)
print 'nu_eff_ff =', "{0:.1f}".format(nu_eff_ff/1.e9), '+/-', 2.4, 'GHz'

nu_eff_CMB1 = np.sum(nu * fnu * nu**(-2.*(1.+al_1)) * nu_be_CMB * dnu)/np.sum(fnu * nu**(-2.*(1.+al_1)) * nu_be_CMB * dnu)
nu_eff_CMB2 = np.sum(nu * fnu * nu**(-2.*(1.+al_2)) * nu_be_CMB * dnu)/np.sum(fnu * nu**(-2.*(1.+al_2)) * nu_be_CMB * dnu)
nu_eff_CMB = 0.5*(nu_eff_CMB1+nu_eff_CMB2)
print 'nu_eff_CMB1 =', "{0:.1f}".format(nu_eff_CMB1/1.e9) #JCH
print 'nu_eff_CMB2 =', "{0:.1f}".format(nu_eff_CMB2/1.e9) #JCH
print 'nu_eff_CMB =', "{0:.1f}".format(nu_eff_CMB/1.e9), '+/-', 2.4, 'GHz'

nu_eff_SZ1 = np.sum(nu * fnu * nu**(-2.*(1.+al_1)) * nu_be_SZ * dnu)/np.sum(fnu * nu**(-2.*(1.+al_1)) * nu_be_SZ * dnu)
nu_eff_SZ2 = np.sum(nu * fnu * nu**(-2.*(1.+al_2)) * nu_be_SZ * dnu)/np.sum(fnu * nu**(-2.*(1.+al_2)) * nu_be_SZ * dnu)
nu_eff_SZ = 0.5*(nu_eff_SZ1+nu_eff_SZ2)
print 'nu_eff_SZ1 =', "{0:.1f}".format(nu_eff_SZ1/1.e9) #JCH
print 'nu_eff_SZ2 =', "{0:.1f}".format(nu_eff_SZ2/1.e9) #JCH
print 'nu_eff_SZ =', "{0:.1f}".format(nu_eff_SZ/1.e9), '+/-', 2.4, 'GHz'

nu_eff_RJ1 = np.sum(nu * fnu * nu**(-2.*(1.+al_1)) * nu**(be_RJ) * dnu)/np.sum(fnu * nu**(-2.*(1.+al_1)) * nu**(be_RJ) * dnu)
nu_eff_RJ2 = np.sum(nu * fnu * nu**(-2.*(1.+al_2)) * nu**(be_RJ) * dnu)/np.sum(fnu * nu**(-2.*(1.+al_2)) * nu**(be_RJ) * dnu)
nu_eff_RJ = 0.5*(nu_eff_RJ1+nu_eff_RJ2)
print 'nu_eff_RJ =', "{0:.1f}".format(nu_eff_RJ/1.e9), '+/-', 2.4, 'GHz'

nu_eff_d1 = np.sum(nu * fnu * nu**(-2.*(1.+al_1)) * nu**(be_d) * dnu)/np.sum(fnu * nu**(-2.*(1.+al_1)) * nu**(be_d) * dnu)
nu_eff_d2 = np.sum(nu * fnu * nu**(-2.*(1.+al_2)) * nu**(be_d) * dnu)/np.sum(fnu * nu**(-2.*(1.+al_2)) * nu**(be_d) * dnu)
nu_eff_d = 0.5*(nu_eff_d1+nu_eff_d2)
print 'nu_eff_dusty =', "{0:.1f}".format(nu_eff_d/1.e9), '+/-', 2.4, 'GHz'
```

```python
### PA2: conversion factors, compact sources ###

al = -0.75  #For ACT, al is near -1
n_samples = 10000
err_eff_f = 2.4 #estimate of uncertainty in effective central freq
OmegaB = np.random.normal(182, 3, n_samples)*1e-9 #for PA2??

smp_nu_eff_CMB = np.random.normal(nu_eff_CMB, err_eff_f, n_samples)
smp_nu_eff_s = np.random.normal(nu_eff_s, err_eff_f, n_samples)
smp_nu_eff_ff = np.random.normal(nu_eff_ff, err_eff_f, n_samples)
smp_nu_eff_d = np.random.normal(nu_eff_d, err_eff_f, n_samples)
smp_nu_eff_RJ = np.random.normal(nu_eff_RJ, err_eff_f, n_samples)

plt.hist(smp_nu_eff_RJ,25)
plt.xlabel('Freq (GHz)')
plt.show()

nu_eff_CMB = smp_nu_eff_CMB*1e9
nu_eff_s = smp_nu_eff_s*1e9
nu_eff_ff = smp_nu_eff_ff*1e9
nu_eff_d = smp_nu_eff_d*1e9
nu_eff_RJ = smp_nu_eff_RJ*1e9
dnu = dnu*1e9
nu = nu*1e9

T_s = np.zeros(10000)
T_ff = np.zeros(10000)
T_RJ = np.zeros(10000)
T_d = np.zeros(10000)

for i in range(10000):
    T_s[i] = c**2/2/kB/OmegaB[i]/(nu_eff_s[i]**be_s)/(nu_eff_RJ[i]**(-2*al)) * np.sum(fnu * nu**(-2*al) * nu**(be_s-2) * dnu)/np.sum(fnu * dnu)*1e-20 

    T_ff[i] = c**2/2/kB/OmegaB[i]/(nu_eff_ff[i]**be_ff)/(nu_eff_RJ[i]**(-2*al)) * np.sum(fnu * nu**(-2*al) * nu**(be_ff-2) * dnu)/np.sum(fnu * dnu)*1e-20 

    T_RJ[i] = c**2/2/kB/OmegaB[i]/(nu_eff_RJ[i]**be_RJ)/(nu_eff_RJ[i]**(-2*al)) * np.sum(fnu * nu**(-2*al) * nu**(be_RJ-2) * dnu)/np.sum(fnu * dnu)*1e-20 

    T_d[i] = c**2/2/kB/OmegaB[i]/(nu_eff_d[i]**be_d)/(nu_eff_RJ[i]**(-2*al)) * np.sum(fnu * nu**(-2*al) * nu**(be_d-2) * dnu)/np.sum(fnu * dnu)*1e-20 

print 'Factor Gamma (microK/Jy):'
print 'synch =', np.mean(T_s), '+/-', np.std(T_s)
print 'free free =', np.mean(T_ff), '+/-', np.std(T_ff)
print 'RJ =', np.mean(T_RJ), '+/-', np.std(T_RJ)
print 'dust =', np.mean(T_d), '+/-', np.std(T_d)
```
